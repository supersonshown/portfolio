#### 1) ë°ì´í„°ì…‹ ë¶ˆëŸ¬ì˜¤ê¸°

```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
import os
import zipfile
import shutil
from pathlib import Path
from IPython.display import display, HTML

# ì§„í–‰ ìƒí™©ì„ í‘œì‹œí•˜ëŠ” í•¨ìˆ˜
def display_progress(message):
    display(HTML(f"<p style='color: #2874A6;'>{message}</p>"))

def merge_datasets(zip_files, output_dir):
    """
    ì—¬ëŸ¬ zip íŒŒì¼ì˜ ë°ì´í„°ì…‹ì„ í•˜ë‚˜ë¡œ ë³‘í•©í•˜ëŠ” í•¨ìˆ˜

    Args:
        zip_files (list): ë³‘í•©í•  zip íŒŒì¼ë“¤ì˜ ê²½ë¡œ ë¦¬ìŠ¤íŠ¸
        output_dir (str): ë³‘í•©ëœ ë°ì´í„°ì…‹ì´ ì €ì¥ë  ê²½ë¡œ
    """
    # ì¶œë ¥ ë””ë ‰í† ë¦¬ ìƒì„±
    os.makedirs(output_dir, exist_ok=True)
    display_progress(f"ì¶œë ¥ ë””ë ‰í† ë¦¬ ìƒì„±: {output_dir}")

    # ê¸°ë³¸ í´ë” êµ¬ì¡° ìƒì„±
    for split in ['train', 'test', 'valid']:
        for subdir in ['images', 'labels']:
            os.makedirs(os.path.join(output_dir, split, subdir), exist_ok=True)
    display_progress("ê¸°ë³¸ í´ë” êµ¬ì¡° ìƒì„± ì™„ë£Œ")

    # ì „ì²´ íŒŒì¼ ì¹´ìš´í„° ì´ˆê¸°í™”
    total_files = 0

    # ê° zip íŒŒì¼ ì²˜ë¦¬
    for zip_path in zip_files:
        display_progress(f"ì²˜ë¦¬ ì¤‘ì¸ íŒŒì¼: {zip_path}")

        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            # ì„ì‹œ ë””ë ‰í† ë¦¬ì— ì••ì¶• í•´ì œ
            temp_dir = os.path.join(output_dir, 'temp')
            os.makedirs(temp_dir, exist_ok=True)
            zip_ref.extractall(temp_dir)

            # íŒŒì¼ ë³‘í•©
            for split in ['train', 'test', 'valid']:
                for subdir in ['images', 'labels']:
                    src_dir = os.path.join(temp_dir, split, subdir)
                    dst_dir = os.path.join(output_dir, split, subdir)

                    # ì†ŒìŠ¤ ë””ë ‰í† ë¦¬ê°€ ì¡´ì¬í•˜ëŠ” ê²½ìš°ì—ë§Œ ì²˜ë¦¬
                    if os.path.exists(src_dir):
                        files_copied = 0
                        for file_name in os.listdir(src_dir):
                            src_path = os.path.join(src_dir, file_name)
                            dst_path = os.path.join(dst_dir, file_name)

                            # íŒŒì¼ì´ ì´ë¯¸ ì¡´ì¬í•˜ëŠ” ê²½ìš° ì²˜ë¦¬
                            if os.path.exists(dst_path):
                                base, ext = os.path.splitext(file_name)
                                counter = 1
                                while os.path.exists(dst_path):
                                    new_name = f"{base}_{counter}{ext}"
                                    dst_path = os.path.join(dst_dir, new_name)
                                    counter += 1

                            shutil.copy2(src_path, dst_path)
                            files_copied += 1
                            total_files += 1

                        if files_copied > 0:
                            display_progress(f"{zip_path}ì˜ {split}/{subdir}ì—ì„œ {files_copied}ê°œ íŒŒì¼ ë³µì‚¬ ì™„ë£Œ")

            # ì„ì‹œ ë””ë ‰í† ë¦¬ ì‚­ì œ
            shutil.rmtree(temp_dir)

    display_progress(f"<b>ë°ì´í„°ì…‹ ë³‘í•© ì™„ë£Œ!</b> ì´ {total_files}ê°œ íŒŒì¼ì´ ì²˜ë¦¬ë˜ì—ˆìŠµë‹ˆë‹¤.")

    # ìµœì¢… í´ë” êµ¬ì¡° í™•ì¸
    def count_files(directory):
        return len([f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))])

    display_progress("<b>ìµœì¢… í´ë” êµ¬ì¡° ë° íŒŒì¼ ìˆ˜:</b>")
    for split in ['train', 'test', 'valid']:
        for subdir in ['images', 'labels']:
            path = os.path.join(output_dir, split, subdir)
            file_count = count_files(path)
            display_progress(f"{split}/{subdir}: {file_count}ê°œ íŒŒì¼")

# ì‹¤í–‰ ì…€
zip_files = [
    r"C:\Users\Qwasz\OneDrive\ë°”íƒ• í™”ë©´\project_4\Face_Recognition_1.zip",
    r"C:\Users\Qwasz\OneDrive\ë°”íƒ• í™”ë©´\project_4\Face_Recognition_2.zip",
    r"C:\Users\Qwasz\OneDrive\ë°”íƒ• í™”ë©´\project_4\project_4_1.v1i.yolov8.zip"
]
output_directory = r"C:\Users\Qwasz\OneDrive\ë°”íƒ• í™”ë©´\project_4\merged_data"

# í•¨ìˆ˜ ì‹¤í–‰
merge_datasets(zip_files, output_directory)
```

#### 1) Label íŒŒì¼ì˜ í´ë˜ìŠ¤ ê°’ ë³€ê²½

```python
import os
from pathlib import Path
from IPython.display import display, HTML

def display_info(message, color="#2874A6", margin_left=0):
    display(HTML(f"<p style='color: {color}; margin-left: {margin_left}px;'>{message}</p>"))

def process_labels_in_directory(labels_dir, split_name):
    """íŠ¹ì • ë””ë ‰í† ë¦¬ì˜ ë¼ë²¨ íŒŒì¼ë“¤ì„ ì²˜ë¦¬í•˜ëŠ” í•¨ìˆ˜"""
    total_files = 0
    modified_files = 0

    # display_info(f"\nğŸ“ Processing {split_name} directory: {labels_dir}", color="#8E44AD")

    # ë””ë ‰í† ë¦¬ ë‚´ì˜ ëª¨ë“  txt íŒŒì¼ ì²˜ë¦¬
    for file_name in os.listdir(labels_dir):
        if file_name.endswith('.txt'):
            total_files += 1
            file_path = os.path.join(labels_dir, file_name)
            modified_lines = []

            # íŒŒì¼ ë‚´ìš© ì½ê¸°
            with open(file_path, 'r') as f:
                lines = f.readlines()

            # íŒŒì¼ëª…ì´ 'me'ë¡œ ì‹œì‘í•˜ëŠ”ì§€ í™•ì¸
            if file_name.startswith('me'):
                target_class = '0'
                # display_info(f"Processing 'me' file: {file_name}", color="#27AE60", margin_left=20)
            else:
                target_class = '1'
                # display_info(f"Processing non-'me' file: {file_name}", color="#E74C3C", margin_left=20)

            # ê° ì¤„ ì²˜ë¦¬
            modified = False
            for line in lines:
                parts = line.strip().split()
                if parts:  # ë¹ˆ ì¤„ì´ ì•„ë‹Œ ê²½ìš°ë§Œ ì²˜ë¦¬
                    if parts[0] != target_class:  # í´ë˜ìŠ¤ IDê°€ ë‹¤ë¥¸ ê²½ìš°ë§Œ ìˆ˜ì •
                        parts[0] = target_class
                        modified = True
                    modified_lines.append(' '.join(parts) + '\n')
                else:
                    modified_lines.append('\n')

            # ë³€ê²½ëœ ë‚´ìš©ì´ ìˆëŠ” ê²½ìš°ë§Œ íŒŒì¼ ë‹¤ì‹œ ì“°ê¸°
            if modified:
                with open(file_path, 'w') as f:
                    f.writelines(modified_lines)
                modified_files += 1

    return total_files, modified_files

# ê¸°ë³¸ ë””ë ‰í† ë¦¬ ê²½ë¡œ
base_dir = r"C:\Users\Qwasz\OneDrive\ë°”íƒ• í™”ë©´\project_4\merged_data"

# ê° split(train/test/valid)ì— ëŒ€í•œ ì²˜ë¦¬
splits = {
    'train': os.path.join(base_dir, 'train', 'labels'),
    'test': os.path.join(base_dir, 'test', 'labels'),
    'valid': os.path.join(base_dir, 'valid', 'labels')
}

# ì „ì²´ í†µê³„ë¥¼ ìœ„í•œ ë³€ìˆ˜
total_stats = {
    'total_files': 0,
    'total_modified': 0
}

# ê° split ì²˜ë¦¬
for split_name, labels_dir in splits.items():
    # display_info(f"\nğŸ”„ Processing {split_name} split...", color="#2E86C1")

    if os.path.exists(labels_dir):
        total_files, modified_files = process_labels_in_directory(labels_dir, split_name)

        # í†µê³„ ì—…ë°ì´íŠ¸
        total_stats['total_files'] += total_files
        total_stats['total_modified'] += modified_files

        # splitë³„ ê²°ê³¼ ì¶œë ¥
        display_info(f"\nğŸ“Š {split_name} ì²˜ë¦¬ ê²°ê³¼:", color="#2E86C1")
        display_info(f"- ì „ì²´ ì²˜ë¦¬ëœ íŒŒì¼: {total_files}ê°œ", margin_left=20)
        display_info(f"- ìˆ˜ì •ëœ íŒŒì¼: {modified_files}ê°œ", margin_left=20)
        display_info(f"- ë³€ê²½ë˜ì§€ ì•Šì€ íŒŒì¼: {total_files - modified_files}ê°œ", margin_left=20)
    else:
        display_info(f"âš ï¸ Warning: {labels_dir} ë””ë ‰í† ë¦¬ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.", color="#F39C12")

# ìµœì¢… ê²°ê³¼ ì¶œë ¥
display_info("\nğŸ“ˆ ì „ì²´ ì²˜ë¦¬ ê²°ê³¼:", color="#2E86C1")
display_info(f"- ì´ ì²˜ë¦¬ëœ íŒŒì¼: {total_stats['total_files']}ê°œ", margin_left=20)
display_info(f"- ì´ ìˆ˜ì •ëœ íŒŒì¼: {total_stats['total_modified']}ê°œ", margin_left=20)
display_info(f"- ì´ ë³€ê²½ë˜ì§€ ì•Šì€ íŒŒì¼: {total_stats['total_files'] - total_stats['total_modified']}ê°œ", margin_left=20)
display_info("\nâœ… ëª¨ë“  ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤!", color="#27AE60")
```

```python
def create_dataset_yaml(dataset_dir, yaml_path, class_names=None, dataset_name="Custom Dataset"):
    """
    ë°ì´í„°ì…‹ì„ ìœ„í•œ YAML íŒŒì¼ì„ ìƒì„±í•˜ëŠ” í•¨ìˆ˜

    Args:
        dataset_dir (str): ë°ì´í„°ì…‹ì˜ root ë””ë ‰í† ë¦¬ ê²½ë¡œ
        yaml_path (str): ìƒì„±í•  YAML íŒŒì¼ì˜ ê²½ë¡œ
        class_names (dict, optional): í´ë˜ìŠ¤ IDì™€ ì´ë¦„ ë§¤í•‘ ë”•ì…”ë„ˆë¦¬
        dataset_name (str): ë°ì´í„°ì…‹ì˜ ì´ë¦„
    """
    display_info(f"ğŸ“ ë°ì´í„°ì…‹ ë””ë ‰í† ë¦¬ ìŠ¤ìº” ì¤‘: {dataset_dir}")

    # í´ë˜ìŠ¤ ì •ë³´ì™€ íŒŒì¼ ìˆ˜ ì¶”ì¶œ
    class_ids, file_counts = scan_labels_directory(dataset_dir)

    # í´ë˜ìŠ¤ ì´ë¦„ ë§¤í•‘
    if class_names is None:
        class_names = {i: f'class_{i}' for i in class_ids}

    # YAML êµ¬ì„±
    yaml_content = {
        'path': os.path.abspath(dataset_dir),
        'train': os.path.join('train', 'images'),
        'val': os.path.join('valid', 'images'),
        'test': os.path.join('test', 'images'),
        'names': class_names,
        'nc': len(class_ids)
    }

    # YAML íŒŒì¼ ì €ì¥
    with open(yaml_path, 'w') as f:
        yaml.safe_dump(yaml_content, f, sort_keys=False, allow_unicode=True)

    # ê²°ê³¼ ì¶œë ¥
    display_info("âœ… YAML íŒŒì¼ ìƒì„± ì™„ë£Œ!")
    display_info(f"ğŸ“ YAML íŒŒì¼ ê²½ë¡œ: {os.path.abspath(yaml_path)}")

    # ë°ì´í„°ì…‹ ì •ë³´ ì¶œë ¥
    display_info("\nğŸ“Š ë°ì´í„°ì…‹ í†µê³„:")
    display_info(f"- ì „ì²´ í´ë˜ìŠ¤ ìˆ˜: {len(class_ids)}", margin_left=20)
    display_info(f"- í•™ìŠµ ë°ì´í„°: {file_counts['train']}ê°œ íŒŒì¼", margin_left=20)
    display_info(f"- ê²€ì¦ ë°ì´í„°: {file_counts['valid']}ê°œ íŒŒì¼", margin_left=20)
    display_info(f"- í…ŒìŠ¤íŠ¸ ë°ì´í„°: {file_counts['test']}ê°œ íŒŒì¼", margin_left=20)

    display_info("\nğŸ·ï¸ í´ë˜ìŠ¤ ì •ë³´:")
    for class_id, class_name in class_names.items():
        display_info(f"- ID {class_id}: {class_name}", margin_left=20)

    return yaml_content

# í´ë˜ìŠ¤ ì´ë¦„ì„ ì§ì ‘ ì§€ì •í•˜ê³  ì‹¶ì€ ê²½ìš° ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” ì˜ˆì‹œ ë”•ì…”ë„ˆë¦¬
example_class_names = {
    0: "my_face",
    1: "other_face",
}
```

```python
# prompt: ê²½ë¡œê°€ '/content/drive/MyDrive/kt/project_4/Datasets/YOLO/merged_data/merged_data' ì¸ ë°ì´í„°ì…‹ì˜ /trian/images ì— .jpg íŒŒì¼ì˜ ì• ë¶€ë¶„ì´ me ì¸ê²ƒê³¼ ì•„ë‹Œê²ƒì˜ ê°œìˆ˜ë¥¼ ì„¸ì¤˜

import os

def count_jpg_files(directory, prefix):
  """
  ì§€ì •ëœ ë””ë ‰í† ë¦¬ì—ì„œ .jpg íŒŒì¼ ì¤‘ prefixë¡œ ì‹œì‘í•˜ëŠ” íŒŒì¼ê³¼ ì•„ë‹Œ íŒŒì¼ì˜ ê°œìˆ˜ë¥¼ ì…‰ë‹ˆë‹¤.

  Args:
    directory: ê²€ìƒ‰í•  ë””ë ‰í† ë¦¬ ê²½ë¡œ
    prefix: íŒŒì¼ ì´ë¦„ ì•ë¶€ë¶„ì— í¬í•¨ë˜ì–´ì•¼ í•  ë¬¸ìì—´

  Returns:
    (prefixë¡œ ì‹œì‘í•˜ëŠ” íŒŒì¼ ê°œìˆ˜, prefixë¡œ ì‹œì‘í•˜ì§€ ì•ŠëŠ” íŒŒì¼ ê°œìˆ˜) íŠœí”Œ
  """
  prefix_count = 0
  non_prefix_count = 0
  for filename in os.listdir(directory):
    if filename.endswith('.jpg'):
      if filename.startswith(prefix):
        prefix_count += 1
      else:
        non_prefix_count += 1
  return prefix_count, non_prefix_count

# ë°ì´í„°ì…‹ ë””ë ‰í† ë¦¬
dataset_dir = '/content/drive/MyDrive/kt/project_4/Datasets/YOLO/merged_data/merged_data/train/images'

# 'me'ë¡œ ì‹œì‘í•˜ëŠ” íŒŒì¼ ê°œìˆ˜ì™€ ì•„ë‹Œ íŒŒì¼ ê°œìˆ˜ë¥¼ ì…‰ë‹ˆë‹¤.
me_count, non_me_count = count_jpg_files(dataset_dir, 'me')

print(f"'me'ë¡œ ì‹œì‘í•˜ëŠ” íŒŒì¼ ê°œìˆ˜: {me_count}")
print(f"'me'ë¡œ ì‹œì‘í•˜ì§€ ì•ŠëŠ” íŒŒì¼ ê°œìˆ˜: {non_me_count}")
```
