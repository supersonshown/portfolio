#### 1) 데이터셋 불러오기

```python
# 필요한 라이브러리 임포트
import os
import zipfile
import shutil
from pathlib import Path
from IPython.display import display, HTML

# 진행 상황을 표시하는 함수
def display_progress(message):
    display(HTML(f"<p style='color: #2874A6;'>{message}</p>"))

def merge_datasets(zip_files, output_dir):
    """
    여러 zip 파일의 데이터셋을 하나로 병합하는 함수

    Args:
        zip_files (list): 병합할 zip 파일들의 경로 리스트
        output_dir (str): 병합된 데이터셋이 저장될 경로
    """
    # 출력 디렉토리 생성
    os.makedirs(output_dir, exist_ok=True)
    display_progress(f"출력 디렉토리 생성: {output_dir}")

    # 기본 폴더 구조 생성
    for split in ['train', 'test', 'valid']:
        for subdir in ['images', 'labels']:
            os.makedirs(os.path.join(output_dir, split, subdir), exist_ok=True)
    display_progress("기본 폴더 구조 생성 완료")

    # 전체 파일 카운터 초기화
    total_files = 0

    # 각 zip 파일 처리
    for zip_path in zip_files:
        display_progress(f"처리 중인 파일: {zip_path}")

        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            # 임시 디렉토리에 압축 해제
            temp_dir = os.path.join(output_dir, 'temp')
            os.makedirs(temp_dir, exist_ok=True)
            zip_ref.extractall(temp_dir)

            # 파일 병합
            for split in ['train', 'test', 'valid']:
                for subdir in ['images', 'labels']:
                    src_dir = os.path.join(temp_dir, split, subdir)
                    dst_dir = os.path.join(output_dir, split, subdir)

                    # 소스 디렉토리가 존재하는 경우에만 처리
                    if os.path.exists(src_dir):
                        files_copied = 0
                        for file_name in os.listdir(src_dir):
                            src_path = os.path.join(src_dir, file_name)
                            dst_path = os.path.join(dst_dir, file_name)

                            # 파일이 이미 존재하는 경우 처리
                            if os.path.exists(dst_path):
                                base, ext = os.path.splitext(file_name)
                                counter = 1
                                while os.path.exists(dst_path):
                                    new_name = f"{base}_{counter}{ext}"
                                    dst_path = os.path.join(dst_dir, new_name)
                                    counter += 1

                            shutil.copy2(src_path, dst_path)
                            files_copied += 1
                            total_files += 1

                        if files_copied > 0:
                            display_progress(f"{zip_path}의 {split}/{subdir}에서 {files_copied}개 파일 복사 완료")

            # 임시 디렉토리 삭제
            shutil.rmtree(temp_dir)

    display_progress(f"<b>데이터셋 병합 완료!</b> 총 {total_files}개 파일이 처리되었습니다.")

    # 최종 폴더 구조 확인
    def count_files(directory):
        return len([f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))])

    display_progress("<b>최종 폴더 구조 및 파일 수:</b>")
    for split in ['train', 'test', 'valid']:
        for subdir in ['images', 'labels']:
            path = os.path.join(output_dir, split, subdir)
            file_count = count_files(path)
            display_progress(f"{split}/{subdir}: {file_count}개 파일")

# 실행 셀
zip_files = [
    r"C:\Users\Qwasz\OneDrive\바탕 화면\project_4\Face_Recognition_1.zip",
    r"C:\Users\Qwasz\OneDrive\바탕 화면\project_4\Face_Recognition_2.zip",
    r"C:\Users\Qwasz\OneDrive\바탕 화면\project_4\project_4_1.v1i.yolov8.zip"
]
output_directory = r"C:\Users\Qwasz\OneDrive\바탕 화면\project_4\merged_data"

# 함수 실행
merge_datasets(zip_files, output_directory)
```

    - Label 파일, 즉 txt 파일의 클래스를 수정하셔야 합니다.

#### 1) Label 파일의 클래스 값 변경

```python
import os
from pathlib import Path
from IPython.display import display, HTML

def display_info(message, color="#2874A6", margin_left=0):
    display(HTML(f"<p style='color: {color}; margin-left: {margin_left}px;'>{message}</p>"))

def process_labels_in_directory(labels_dir, split_name):
    """특정 디렉토리의 라벨 파일들을 처리하는 함수"""
    total_files = 0
    modified_files = 0

    # display_info(f"\n📁 Processing {split_name} directory: {labels_dir}", color="#8E44AD")

    # 디렉토리 내의 모든 txt 파일 처리
    for file_name in os.listdir(labels_dir):
        if file_name.endswith('.txt'):
            total_files += 1
            file_path = os.path.join(labels_dir, file_name)
            modified_lines = []

            # 파일 내용 읽기
            with open(file_path, 'r') as f:
                lines = f.readlines()

            # 파일명이 'me'로 시작하는지 확인
            if file_name.startswith('me'):
                target_class = '0'
                # display_info(f"Processing 'me' file: {file_name}", color="#27AE60", margin_left=20)
            else:
                target_class = '1'
                # display_info(f"Processing non-'me' file: {file_name}", color="#E74C3C", margin_left=20)

            # 각 줄 처리
            modified = False
            for line in lines:
                parts = line.strip().split()
                if parts:  # 빈 줄이 아닌 경우만 처리
                    if parts[0] != target_class:  # 클래스 ID가 다른 경우만 수정
                        parts[0] = target_class
                        modified = True
                    modified_lines.append(' '.join(parts) + '\n')
                else:
                    modified_lines.append('\n')

            # 변경된 내용이 있는 경우만 파일 다시 쓰기
            if modified:
                with open(file_path, 'w') as f:
                    f.writelines(modified_lines)
                modified_files += 1

    return total_files, modified_files

# 기본 디렉토리 경로
base_dir = r"C:\Users\Qwasz\OneDrive\바탕 화면\project_4\merged_data"

# 각 split(train/test/valid)에 대한 처리
splits = {
    'train': os.path.join(base_dir, 'train', 'labels'),
    'test': os.path.join(base_dir, 'test', 'labels'),
    'valid': os.path.join(base_dir, 'valid', 'labels')
}

# 전체 통계를 위한 변수
total_stats = {
    'total_files': 0,
    'total_modified': 0
}

# 각 split 처리
for split_name, labels_dir in splits.items():
    # display_info(f"\n🔄 Processing {split_name} split...", color="#2E86C1")

    if os.path.exists(labels_dir):
        total_files, modified_files = process_labels_in_directory(labels_dir, split_name)

        # 통계 업데이트
        total_stats['total_files'] += total_files
        total_stats['total_modified'] += modified_files

        # split별 결과 출력
        display_info(f"\n📊 {split_name} 처리 결과:", color="#2E86C1")
        display_info(f"- 전체 처리된 파일: {total_files}개", margin_left=20)
        display_info(f"- 수정된 파일: {modified_files}개", margin_left=20)
        display_info(f"- 변경되지 않은 파일: {total_files - modified_files}개", margin_left=20)
    else:
        display_info(f"⚠️ Warning: {labels_dir} 디렉토리를 찾을 수 없습니다.", color="#F39C12")

# 최종 결과 출력
display_info("\n📈 전체 처리 결과:", color="#2E86C1")
display_info(f"- 총 처리된 파일: {total_stats['total_files']}개", margin_left=20)
display_info(f"- 총 수정된 파일: {total_stats['total_modified']}개", margin_left=20)
display_info(f"- 총 변경되지 않은 파일: {total_stats['total_files'] - total_stats['total_modified']}개", margin_left=20)
display_info("\n✅ 모든 작업이 완료되었습니다!", color="#27AE60")
```

```python
# prompt: 경로가 '/content/drive/MyDrive/kt/project_4/Datasets/YOLO/merged_data/merged_data' 인 데이터셋의 /trian/images 에 .jpg 파일의 앞 부분이 me 인것과 아닌것의 개수를 세줘

import os

def count_jpg_files(directory, prefix):
  """
  지정된 디렉토리에서 .jpg 파일 중 prefix로 시작하는 파일과 아닌 파일의 개수를 셉니다.

  Args:
    directory: 검색할 디렉토리 경로
    prefix: 파일 이름 앞부분에 포함되어야 할 문자열

  Returns:
    (prefix로 시작하는 파일 개수, prefix로 시작하지 않는 파일 개수) 튜플
  """
  prefix_count = 0
  non_prefix_count = 0
  for filename in os.listdir(directory):
    if filename.endswith('.jpg'):
      if filename.startswith(prefix):
        prefix_count += 1
      else:
        non_prefix_count += 1
  return prefix_count, non_prefix_count

# 데이터셋 디렉토리
dataset_dir = '/content/drive/MyDrive/kt/project_4/Datasets/YOLO/merged_data/merged_data/train/images'

# 'me'로 시작하는 파일 개수와 아닌 파일 개수를 셉니다.
me_count, non_me_count = count_jpg_files(dataset_dir, 'me')

print(f"'me'로 시작하는 파일 개수: {me_count}")
print(f"'me'로 시작하지 않는 파일 개수: {non_me_count}")
```

```python
# prompt: dataset_dir = '/content/drive/MyDrive/kt/project_4/Datasets/YOLO/merged_data/merged_data/train/images'
# 에 있는 .jpg 파일의 압부분이 me 가 아니 파일의3465개의 파일을 랜덤으로 삭제하고 그 삭제한 .jpg 파일과 같은 이름의 .txt 파일을 /content/drive/MyDrive/kt/project_4/Datasets/YOLO/merged_data/merged_data/train/labels 에서도 삭제해줘

import os
import random
import shutil

dataset_dir = '/content/drive/MyDrive/kt/project_4/Datasets/YOLO/merged_data/merged_data/train/images'
label_dir = '/content/drive/MyDrive/kt/project_4/Datasets/YOLO/merged_data/merged_data/train/labels'

jpg_files = [f for f in os.listdir(dataset_dir) if f.endswith('.jpg') and not f.startswith('me')]
num_files_to_delete = min(3465, len(jpg_files))

if num_files_to_delete > 0:
    files_to_delete = random.sample(jpg_files, num_files_to_delete)

    for filename in files_to_delete:
        jpg_path = os.path.join(dataset_dir, filename)
        txt_filename = filename[:-4] + '.txt'
        txt_path = os.path.join(label_dir, txt_filename)

        try:
            os.remove(jpg_path)
            if os.path.exists(txt_path):
                os.remove(txt_path)
            print(f"Deleted: {jpg_path}, {txt_path}")
        except OSError as e:
            print(f"Error deleting {jpg_path}: {e}")
else:
    print("No files to delete.")
```
